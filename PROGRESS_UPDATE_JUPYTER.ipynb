## ğŸ“ˆ **Recent User Activity Analysis**

```python
# Analyze recent user activity
if db_file.exists():
    conn = sqlite3.connect(db_file)
    
    # Get recent quotes
    query = """
    SELECT provider, gpu_type, price_hr, region, ts, spot, selected
    FROM quotes 
    ORDER BY ts DESC
    """
    
    df_quotes = pd.read_sql_query(query, conn)
    
    if not df_quotes.empty:
        print("ğŸ“Š Recent Quote Activity")
        print("=" * 40)
        
        for i, row in df_quotes.iterrows():
            timestamp = pd.to_datetime(row['ts'])
            time_str = timestamp.strftime('%Y-%m-%d %H:%M:%S')
            spot_type = 'spot' if row['spot'] else 'on-demand'
            selected_str = 'âœ… SELECTED' if row['selected'] else ''
            
            print(f"\nQuote {i+1}:")
            print(f"  Provider: {row['provider']}")
            print(f"  GPU: {row['gpu_type']}")
            print(f"  Price: ${row['price_hr']}/hr")
            print(f"  Region: {row['region']}")
            print(f"  Type: {spot_type}")
            print(f"  Time: {time_str}")
            print(f"  Selected: {selected_str}")
        
        # Analysis
        print(f"\nğŸ¯ User Behavior Analysis:")
        print(f"   Provider preference: {df_quotes['provider'].iloc[0]} (all quotes)")
        print(f"   GPU choice: {df_quotes['gpu_type'].iloc[0]}")
        print(f"   Price range: ${df_quotes['price_hr'].min():.2f} - ${df_quotes['price_hr'].max():.2f}/hr")
        print(f"   Best value: ${df_quotes[df_quotes['selected'] == True]['price_hr'].iloc[0]:.2f}/hr (selected)")
        print(f"   Instance type: {'Spot' if df_quotes[df_quotes['selected'] == True]['spot'].iloc[0] else 'On-demand'} (selected)")
        
    else:
        print("âŒ No quotes found")
    
    conn.close()
else:
    print("âŒ No database found")
```
